```python
import pandas as pd
import numpy as np

df = pd.read_csv("2017_jun_final.csv")
```

```python
print("Розмір таблиці", df.shape)
```

```python
print("Типи стовпців", df.dtypes)
```

```python
skip_percent = (df.isnull().sum() / len(df) * 100).round(2)
print("Частка пропусків у відсотках:\n", skip_percent)
```

```python
print(df.columns)
del_col = [col for col in df.columns if df[col].isnull().any() and col != "Мова програмування"]
df_cleaned = df.drop(columns=del_col)
print(df_cleaned.columns)
```

```python
skip_percent_clean = (df_cleaned.isnull().sum() / len(df_cleaned) * 100).round(2)
print(skip_percent_clean)
```

```python
df_drop_na = df.dropna()
df_drop_na
```

```python
print("\nРозмір таблиці після видалення рядків із пропусками:", df_drop_na.shape)
```

```python
python_data = df[df["Мова програмування"].str.contains("Python", case=False, na=False)]
print(python_data.head())
```

```python
print("Розмір таблиці", python_data.shape)
```

```python
grouped_python_data_position = python_data.groupby("Посада")
print(grouped_python_data_position.head())
```

```python
salary_stats = grouped_python_data_position["Зарплата на місяць"].agg(["min", "max"]).reset_index()
print(salary_stats)
```

```python
def fill_avg_salary(row):
    return round((row["min"] + row["max"]) / 2, 2)
```

```python
salary_stats["avg"] = salary_stats.apply(fill_avg_salary, axis=1)
print(salary_stats.head())
```

```python
print(salary_stats["avg"].describe())
```

```python
salary_stats.to_csv("python_salary.csv", index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.barplot(
    data=salary_stats.melt(id_vars="Посада", value_vars=["min", "max"], var_name="Тип", value_name="Зарплата"),
    x="Зарплата", y="Посада", hue="Тип", palette="crest"
)

plt.title("Мінімальні та максимальні зарплати за посадою серед Python-розробників", fontsize=14)
plt.xlabel("Зарплата на місяць $")
plt.ylabel("Посада")
plt.grid(alpha=0.3, zorder=0)
sns.despine()
plt.legend(title="Тип зарплати")
plt.show()
```

```python
avg_salary_by_level = (
    python_data.groupby("Посада")["Зарплата на місяць"]
    .mean()
    .sort_values(ascending=False)
    .reset_index()
)
```

```python
plt.figure(figsize=(8, 5))

sns.barplot(
    data=avg_salary_by_level,
    x="Посада",
    y="Зарплата на місяць",
    palette="viridis",
    zorder = 2
)
plt.title("Середня зарплата Python-розробників за посадою", fontsize=14)
plt.xlabel("Посада")
plt.ylabel("Середня зарплата $")
plt.xticks(rotation=45)
plt.grid(alpha=0.3, zorder = 1)
sns.despine()
plt.show()
```

```python
plt.figure(figsize=(8, 5))

sns.countplot(
    data=python_data,
    y="Посада",
    order=python_data["Посада"].value_counts().index,
    palette="pastel"
)
plt.title("Кількість Python-розробників за посадою", fontsize=14)
plt.xlabel("Кількість осіб")
plt.ylabel("Посада")
plt.grid(alpha=0.3, zorder=0)
sns.despine()
plt.show()
```
